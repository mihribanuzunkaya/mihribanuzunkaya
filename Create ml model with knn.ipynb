```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# STEP 2 : IMPORT / EXPLORE DATASET

# STEP 3 : DATA CLEANING

```python
training_set.drop(['Participant', 'Anim Time to First Appearance [ms]', 'Anim Appearance Count', 'Anim Visible Time [ms]', 'SI Time to First Appearance [ms]', 'SI Appearance Count', 'SI Visible Time [ms]' ], axis=1, inplace= True)
```

```python
training_set
```

```python
pd.get_dummies(training_set['Class'])
```

```python
TD = pd.get_dummies(training_set['Class'], drop_first= True )
```

```python
TD
```

```python
training_set.drop(['Class'], axis =1 , inplace=True)
```

```python
training_set= pd.concat([training_set, TD], axis=1 )
```

```python
training_set
```

```python
X = training_set.drop(['TD'], axis=1)
```

```python
X
```

```python
y = training_set['TD']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split (X, y, test_size = 0.35)
```

```python
X_train.shape
```

```python
X_train
```

# STEP 4 : TRAINING MODEL 

```python
# Fitting K-NN to the Training set
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix

classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

# STEP 5 : EVALUATING THE MODEL

```python
y_predict = classifier.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap (cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

# CROSS VALIDATION

```python
# importing cross_val_score function
from sklearn.model_selection import cross_val_score
# 10-fold CV
scores = cross_val_score(classifier, X, y, scoring = 'f1', cv=10 )
scores
```

```python
print (np.mean(scores)) 
```

```python
# 10_fold CV on test data
scores_test = cross_val_score(classifier, X_test, y_test, cv=10)
scores_test
```

```python
print(np.mean(scores_test))
```

```python
from sklearn.model_selection import KFold
kfold = KFold (n_splits= 3, shuffle = True, random_state= 0)  
score = cross_val_score(classifier, X, y,  cv=kfold)
score.mean()

```
